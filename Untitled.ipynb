```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
import pandas as pd


df = pd.read_csv(r"C:\Users\abc\Desktop\project data\train.csv")
print(df.head())
print(df.info())
print(df.describe())

```

```python
for col in df.columns:
    if 'date' in col.lower():
        df[col] = pd.to_datetime(df[col], dayfirst=True)
        print("Converted:", col)
```

```python
import pandas as pd

# Data Cleaning

try:
    # Remove extra spaces in column names (very important)
    df.columns = df.columns.str.strip()

    # Check if column exists before converting
    if 'Order_Date' in df.columns:
        # Convert date column (DD/MM/YYYY format)
        df['Order_Date'] = pd.to_datetime(
            df['Order_Date'],
            dayfirst=True,
            errors='coerce'
        )
        print("Order_Date column converted successfully ✅")
    else:
        print("Column 'Order_Date' not found ❌")

    # Check missing values
    print("\nMissing Values:\n")
    print(df.isnull().sum())

except Exception as e:
    print("An error occurred:", e)
```

```python
# total revenue
total_revenue = df['Sales'].sum()
print("Total Revenue:", total_revenue)
```

```python
#Revenue by Region


region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

print(region_sales)

region_sales.plot(kind='bar')
plt.title("Revenue by Region")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create Top 10 Products
top_products = (
    df.groupby('Product Name')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# Print result
print(top_products)

# Plot
top_products.plot(kind='barh', figsize=(8,6))
plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.gca().invert_yaxis()  # Highest at top
plt.show()
```

```python


# Make sure column names have no extra spaces
df.columns = df.columns.str.strip()

# Convert date column (your dataset has "Order Date")
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')

# Create Month column
df['Month'] = df['Order Date'].dt.to_period('M')

# Group by Month
monthly_sales = df.groupby('Month')['Sales'].sum()

# Plot
monthly_sales.plot(figsize=(10,5))

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Top Customers (Customer Lifetime Value)

top_customers = (
    df.groupby('Customer Name')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_customers)
```

```python
# Calculate total revenue
total_revenue = df['Sales'].sum()

# Group by Category
category_sales = df.groupby('Category')['Sales'].sum()

# Calculate percentage contribution
percentage = (category_sales / total_revenue) * 100

# Create final table
category_contribution = pd.DataFrame({
    'Revenue': category_sales,
    'Percentage (%)': percentage
})

print(category_contribution)
```

```python


# Correlation matrix (numeric columns only)
corr = df.corr(numeric_only=True)

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Correlation Matrix")
plt.show()
```
